# Chains

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm


N = 1000

x = np.random.normal(size=N)
z = np.random.normal(x)
y = np.random.normal(z)

df = pd.DataFrame({'x': x, 'y': y, 'z': z})
```

```python
pd.plotting.scatter_matrix(df, figsize=(6,6));
```

```python
model = sm.OLS(exog=df['x'], endog=df['y'])
result = model.fit()
result.summary()
```

```python
model = sm.OLS(exog=df['z'], endog=df['y'])
result = model.fit()
result.summary()
```

```python
model = sm.OLS(exog=df[['z', 'x']], endog=df['y'])
result = model.fit()
result.summary()
```

# Forks

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm


N = 1000

z = np.random.normal(size=N)
x = np.random.normal(z)
y = np.random.normal(z)

df = pd.DataFrame({'x': x, 'y': y, 'z': z})

pd.plotting.scatter_matrix(df, figsize=(6,6));
```

```python
model = sm.OLS(exog=df[['z', 'x']], endog=df['y'])
result = model.fit()
result.summary()
```

## Colliders

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm


N = 2000

x = np.random.normal(size=N)
y = np.random.normal(size=N)
z = 1.*np.logical_or(x > 0, y > 0)


df = pd.DataFrame({'x': x, 'y': y, 'z': z})
students_only = df[df.z == 1]

pd.plotting.scatter_matrix(df, figsize=(6,6));
```

```python
pd.plotting.scatter_matrix(students_only, figsize=(6,6));
```

```python
students_only.corr()
```

```python
model = sm.OLS(exog=df['x'], endog=df['y'])
result = model.fit()
result.summary()
```

```python
model = sm.OLS(exog=df[['x', 'z']], endog=df['y'])
result = model.fit()
result.summary()
```
