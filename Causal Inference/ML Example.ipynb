```python
import numpy as np
import pandas as pd
```

```python
N = 10000

x = np.random.normal(size=N)

y = np.random.normal(x)
c = np.random.normal(y)
s = np.random.binomial(1, p=1. / (1. + np.exp(-c)))

df = pd.DataFrame({'X': x, 'Y': y, 'C': c, 
                   'S': s, 'p_true': 1. / (1. + np.exp(-c))})
```

```python
from statsmodels.api import OLS


df['intercept'] = 1

model = OLS(endog=df['Y'],
            exog=df[['X', 'intercept']])
result = model.fit()
result.summary()
```

```python
from statsmodels.api import OLS


df['intercept'] = 1

model = OLS(endog=df[df.S==1]['Y'],
            exog=df[df.S==1][['X', 'intercept']])
result = model.fit()
result.summary()
```

```python
from statsmodels.api import Logit


df['intercept'] = 1

model = Logit(endog=df['S'],
              exog=df[['C', 'X', 'intercept']])
result = model.fit()
result.summary()
```

```python
df['p'] = result.predict()
```

```python
df['weight'] = 1 / df['p']
```

```python
from statsmodels.api import WLS


df['intercept'] = 1

model = WLS(endog=df[df.S==1]['Y'],
            exog=df[df.S==1][['X', 'intercept']],
            weights=df[df.S==1]['weight'])
result = model.fit()
result.summary()
```
