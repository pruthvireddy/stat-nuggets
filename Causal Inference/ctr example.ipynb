```python
import numpy as np
import pandas as pd



N = 100000

q = np.random.binomial(1, p=0.5, size=N) 
x = (1-q) * np.random.choice(range(20), size=N) + \
       q  * np.floor(np.random.normal(13, size=N))

y = np.random.beta(5 + 2*q , 495)

df = pd.DataFrame({'X': x, 
                   'Y': y, 
                   'Q': q})
```

```python
import seaborn as sns
import matplotlib.pyplot as pp
```

## Plot $E[Y|X=x]$

```python
sns.lineplot(data=df, x='X', y='Y'); 
pp.xlim(5, 20); 
pp.ylim(0, 0.02)
pp.ylabel('$E[Y|X=x]$')
pp.xlabel('$x$')
```

We can see there's a clear dependence between $Y$ and $X$.

## Plot $E[Y|Q=q, X=x]$

```python
sns.lineplot(data=df, x='X', y='Y', hue='Q'); 
pp.xlim(5, 20); 
pp.ylim(0, 0.02)
pp.ylabel('$E[Y|X=x, Q=q]$')
pp.xlabel('$x$')
```

So now we can see (as you could read from the DGP) that there's no statistical dependence between $X$ and $Y$ within either $Q=q$ group! Which is the right answer? If we intervene to set title length around 13, is there an effect or not?
